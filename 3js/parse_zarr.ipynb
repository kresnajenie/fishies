```python
def normalize(data, new_min=-10, new_max=10):
    # Calculate the current min and max values of the x_spatial column
    min_value = data.min()
    max_value = data.max()

    # Normalize the x_spatial column to be between -10 and 10
    return ((data - min_value) / (max_value - min_value)) * (new_max - new_min) + new_min
```

```python
def parse_json(adataa, filename, color_dict, group="MERFISH celltype", threed=False):
    adata = adataa.copy()
    df = adata.to_df()
    df['celltype'] = adata.obs[group]

    # df = df.iloc[:,-2:]

    df["x_spatial"] = adata.obsm["X_spatial"][:,0].tolist()
    df["y_spatial"] = adata.obsm["X_spatial"][:,1].tolist()


    if threed:
        sc.tl.umap(adata, n_components=3)
        df["z_umap"] = adata.obsm["X_umap"][:,2].tolist()
        df['z_umap_norm'] = normalize(df['z_umap'])

    df["x_umap"] = adata.obsm["X_umap"][:,0].tolist()
    df["y_umap"] = adata.obsm["X_umap"][:,1].tolist()

    df['x_umap_norm'] = normalize(df['x_umap'])
    df['y_umap_norm'] = normalize(df['y_umap'])

    # df = df.iloc[:, 1:]

    df['color'] = df['celltype'].map(color_dict)

    df['x_spatial_norm'] = normalize(df['x_spatial'])
    df['y_spatial_norm'] = normalize(df['y_spatial'])


    return df

    # Convert the DataFrame to JSON with 'records' orientation
    json_data = df.to_json(orient='records')

    # Specify the file path where you want to save the JSON afile
    file_path = f"{filename}"

    # Open the file in write mode and write the JSON data to it
    with open(file_path, 'w') as json_file:
        json_file.write(json_data)

    print(f'JSON data saved to {file_path}')

```
